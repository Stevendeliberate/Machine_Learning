```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#输入数据
X = np.array([[1,3,3],
              [1,4,3],
              [1,1,1],
              [1,0,2]])
#标签
Y = np.array([[1],
              [1],
              [-1],
              [-1]])

#权值初始化，3行1列，取值范围-1到1
W = (np.random.random([3,1])-0.5)*2

plt.scatter(X[:2,1] , X[:2,2] , c = 'b')
plt.scatter(X[2:4,1] , X[2:4,2] , c = 'r')
a = np.array([0,5])
plt.plot(a,-W[0]/W[2]- a * W[1] / W[2] ,'yellow')
plt.show()
print(-W[0]/W[2])
print(-W[1] / W[2])
```

```python
def update(x , y , lr , epoches):
    #初始化数据
    xMat = np.mat(x)
    yMat = np.mat(y)
    theta = np.mat(np.ones([xMat.shape[1],1]))
    thetaList = theta
    #进行迭代更新theta
    for i in range(epoches):
        theta_gard = np.mat(np.sign(xMat * theta))
        if (theta_gard == yMat).all():
            return theta,thetaList
        theta = theta + xMat.T * (yMat - theta_gard) * lr / xMat.shape[0]
        thetaList = np.c_[thetaList , theta]
    return theta,thetaList
```

```python
W , thetaList = update(X,Y,0.01,1000)
thetaList.shape
```

```python
plt.scatter(X[:2,1] , X[:2,2] , c = 'b')
plt.scatter(X[2:4,1] , X[2:4,2] , c = 'r')
a = np.array([0,5])
plt.plot(a,-W[0,0]/W[2,0]- a * W[1,0] / W[2,0] ,'yellow')
plt.show()
print(-W[0,0]/W[2,0])
print(-W[1,0] / W[2,0])
```

```python
w = thetaList.T
for i in range(w.shape[0]):
    if i% 10 == 0:
        plt.scatter(X[:2,1] , X[:2,2] , c = 'b')
        plt.scatter(X[2:4,1] , X[2:4,2] , c = 'r')
        a = np.array([0,5])
        plt.plot(a,-w[i,0]/w[i,2]- a * w[i,1] / w[i,2] ,'yellow')
        print(w[i])
        plt.show()
```
