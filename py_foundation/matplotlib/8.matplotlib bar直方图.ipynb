```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y = 2**x + 10
plt.bar(x,y)
plt.show()
```

```python
plt.bar(x,-y)
plt.show()
```

```python
plt.bar(x,y,facecolor = '#9999ff',edgecolor = 'red')
plt.show()
```

```python
plt.bar(x,y,facecolor = '#9999ff',edgecolor = 'red')
for x,y in zip(x,y):
    plt.text(x,y,'%.2f'%y,ha='center',va = 'bottom')
plt.show()
```
