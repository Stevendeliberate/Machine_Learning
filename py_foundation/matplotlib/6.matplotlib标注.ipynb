```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1,1,100)
y1 = x*3 + 1
y2 = x**2

new_ticks = np.linspace(-2,2,11)
plt.figure()
plt.xlabel('A B')
plt.ylabel('C D')

l1, = plt.plot(x,y1,color = 'red',linewidth = 1.0,linestyle = '--')
l2, = plt.plot(x,y2,color = 'yellow',linewidth = 2.0,linestyle = '-')

plt.legend(handles = [l1,l2],labels = ['test1','tes2'],loc = 'best')

plt.xticks(new_ticks)
plt.yticks([-1.0,0,1,2,3],['level 1','level 2','level 3','level 4'])

#修改坐标轴
#gca => get current axis
ax =plt.gca()

#把右边和上边的边框去掉 也可更改为其他颜色
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

#把x轴的刻度设置为‘bottom’
#把y轴的刻度设置为‘left’
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
#设置bottom对应到0点
#设置left对应到0点
#可用于滑动坐标轴，改变零点所在位置
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

x0 = 0.5
y0 = 3*x0 + 1
#画点
plt.scatter(x0,y0,color = 'b')

#画虚线
plt.plot([x0,x0],[y0,0],linestyle = '--',linewidth = 5)

plt.annotate(r'$2x +1 = %s$'%y0,xy = (x0,y0),xytext = (+30,-30),textcoords = 'offset points',fontsize = 16,
            arrowprops = dict(arrowstyle = '->',connectionstyle = 'arc3,rad = 0.2'))

plt.text(-1,2,r'$this\ is\ text$',fontdict = {'size':'16','color':'r'})
plt.show()
```
