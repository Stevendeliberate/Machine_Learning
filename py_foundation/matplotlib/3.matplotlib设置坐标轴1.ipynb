```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1,1,100)
y1 = x*3 + 1
y2 = x**2

plt.figure()
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
plt.figure()#设置颜色。线宽，线的风格
plt.plot(x,y1,color = 'red',linewidth = 1.0,linestyle = '--')
plt.plot(x,y2,color = 'yellow',linewidth = 2.0,linestyle = '-')
```

```python
#设置坐标轴显示的范围
plt.figure()
plt.xlim((-1,3))
plt.ylim((-1,1))
plt.plot(x,y1,color = 'red',linewidth = 1.0,linestyle = '--')
plt.plot(x,y2,color = 'yellow',linewidth = 2.0,linestyle = '-')

plt.figure()
plt.plot(x,y1,color = 'red',linewidth = 1.0,linestyle = '--')
plt.plot(x,y2,color = 'yellow',linewidth = 2.0,linestyle = '-')

plt.show()
```

```python
#设置坐标轴lable的名称
plt.figure()
plt.xlabel('A B')
plt.ylabel('C D')
plt.plot(x,y1,color = 'red',linewidth = 1.0,linestyle = '--')
plt.plot(x,y2,color = 'yellow',linewidth = 2.0,linestyle = '-')

plt.figure()
plt.plot(x,y1,color = 'red',linewidth = 1.0,linestyle = '--')
plt.plot(x,y2,color = 'yellow',linewidth = 2.0,linestyle = '-')

plt.show()
```
