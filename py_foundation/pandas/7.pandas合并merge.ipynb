```python
import pandas as pd
```

```python
left = pd.DataFrame({'key':['K0','K1','K2','K3']
                   ,'A':['A0','A1','A2','A3']
                   ,'B':['B0','B1','B2','B3']})
right = pd.DataFrame({'key':['K0','K1','K2','K3']
                   ,'C':['C0','C1','C2','C3']
                   ,'D':['D0','D1','D2','D3']})
print(left)
print(right)
```

```python
res = pd.merge(left,right,on='key')
res
```

```python
left1 = pd.DataFrame({'key':['K0','K1','K2','K3']
                   ,'A':['A0','A1','A2','A3']
                   ,'B':['B0','B1','B2','B3']
                   ,'key1':['K1','K2','K30','K40']})
right1 = pd.DataFrame({'key':['K0','K1','K2','K3']
                   ,'C':['C0','C1','C2','C3']
                   ,'D':['D0','D1','D2','D3']
                   ,'key1':['K1','K20','K3','K40']})
print(left1)

print(right1)
```

```python
#how = ['left','right','inner','outer']
res1 = pd.merge(left1,right1,on = ['key1','key'],how = 'outer') #how 默认 inner   outer是指 所有列合并 不存在的列用NaN填充
res1
```

```python
res2 = pd.merge(left1,right1,on = ['key1','key'],how = 'inner') #必须两个df都有的列 才能合并
res2
```

```python
res3 = pd.merge(left1,right1,on = ['key1','key'],how = 'left') #合并时只考虑左边的df 不考虑右边的df
res3
```

```python
res3 = pd.merge(left1,right1,on = ['key1','key'],how = 'outer',indicator='indicator_column') 
res3
```

```python
left = pd.DataFrame({'A':['A0','A1','A2']
                   ,'B':['B0','B1','B2']}
                   ,index = ['K0','K1','K2'])
right = pd.DataFrame({'C':['C0','C1','C2']
                   ,'D':['D0','D1','D2']}
                   ,index = ['K0','K2','K3'])
print(left)
print(right)
```

```python
res = pd.merge(left,right,left_index = True,right_index = True,how = 'outer')
res    
```

```python
boys = pd.DataFrame({'k':['K0','K1','K2'],'age':[1,2,3]})
girls = pd.DataFrame({'k':['K0','K0','K3'],'age':[4,5,6]})
print(boys)
print(girls)
```

```python
res = pd.merge(boys,girls,on='k',suffixes=['_boy','_girl'],how = 'outer')
res
```
